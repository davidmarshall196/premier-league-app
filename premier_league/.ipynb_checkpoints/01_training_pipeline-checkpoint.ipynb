```python
regressor_1 = training.train_model(
    training_data.drop(['FTR', 'FTHG','FTAG'], axis=1),
    training_data['FTHG'],
    classification=False,
    verbose=False
)
```

```python
joblib.dump(regressor_1, constants.home_model_name)
```

###Predict Home Goals
