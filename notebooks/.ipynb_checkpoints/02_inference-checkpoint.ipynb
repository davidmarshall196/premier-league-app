```python
# Shap values
explainer = shap.TreeExplainer(regressor_1)
shap_values = explainer.shap_values(transformed_data[regressor_1.feature_names_])
```

```python
shap.summary_plot(
        shap_values, 
        transformed_data[regressor_1.feature_names_],
        plot_type = 'bar'
)
```

```python
class_names = ['Home win', 'Draw', 'Away win']
# Loop through each class and create a summary plot
for i in range(len(shap_values)):
    shap.summary_plot(
        shap_values[i], 
        transformed_data[regressor_1.feature_names_], 
        show=False,
        plot_type='bar'
    )
    plt.title(class_names[i])
    plt.show()
```

```python
shap.dependence_plot(
    'DiffPts', 
    shap_values[0], 
    transformed_data[regressor_1.feature_names_]
)
```

```python
shap.dependence_plot(
    'HomeTeam', 
    shap_values[0], 
    transformed_data[classifier.feature_names_]
)
```
